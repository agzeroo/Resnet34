```python
import os
import shutil
# from bing_image_downloader.bing_image_downloader import downloader


directory_list = [
    'C:/Users/master15/Desktop/0812resnet/train/',
    'C:/Users/master15/Desktop/0812resnet/test/',
]
# 초기 디렉토리 만들기
for directory in directory_list:
    if not os.path.isdir(directory):
        os.makedirs(directory)


# 수집한 이미지를 학습 데이터와 평가 데이터로 구분하는 함수
def dataset_split(query, train_cnt):
    # 학습 및 평가 데이터셋 디렉토리 만들기
    for directory in directory_list:
        if not os.path.isdir(directory + '/' + query):
            os.makedirs(directory + '/' + query)
    # 학습 및 평가 데이터셋 준비하기
    cnt = 0
    for file_name in os.listdir(query):
        if cnt < train_cnt:
            print(f'[Train Dataset] {file_name}')
            shutil.move(query + '/' + file_name, 'C:/Users/master15/Desktop/0812resnet/train/' + query + '/' + file_name)
        else:
            print(f'[Test Dataset] {file_name}')
            shutil.move(query + '/' + file_name, 'C:/Users/master15/Desktop/0812resnet/test/' + query + '/' + file_name)
        cnt += 1
    shutil.rmtree(query)
```

```python
query = 'ShortNarrow_0'
dataset_split(query, 810)
```

```python
query = 'LongNarrow_1'
dataset_split(query, 200)
```

```python
query = 'LongWide_2'
dataset_split(query, 500)
```
